```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import imageio
```

```python
def LoG (x,y,sigma):
    return -(1-(x**2+y**2)/(2*sigma**2)) * np.exp(-(x**2+y**2)/(2*sigma**2)) /(np.pi*sigma**4)
```

```python
def get_LoG_mask(size,sigma):
    laplace=np.zeros((size,size))
    for row in range(size):
        for column in range(size):
            laplace[row,column]=LoG(column-np.floor(size/2),np.floor(size/2)-row,sigma)
        
    laplace=laplace/np.sum(laplace)
    return laplace
```

```python
laplace=get_LoG_mask(13,3)
laplace
```

```python
laplace=(laplace-laplace.min())*255/laplace.max()
```

```python
plt.imshow(laplace,cmap=plt.get_cmap('gray'))
```

```python
ax = plt.axes(projection='3d')
X,Y=np.meshgrid(range(-2,3),range(-2,3))
Z=LoG(X,Y,1)
ax.plot_surface(X,Y ,Z, rstride=1, cstride=1,
                cmap='jet', edgecolor='none')
ax.set_title('surface');
```

```python
img=imageio.imread("sample.png",pilmode='L')
```

```python
plt.imshow(img,cmap=plt.get_cmap('gray'))
```

```python
laplace=get_LoG_mask(5,3)
height = img.shape[0]
width = img.shape[1]
img_out=np.zeros((height,width))

for i in np.arange(2, height-2):
    for j in np.arange(2, width-2):        
        sum = 0
        for k in np.arange(-2, 3):
            for l in np.arange(-2, 3):
                a = img.item(i+k, j+l)
                w = laplace[2+k, 2+l]
                sum = sum + (w * a)
        b = sum
        img_out.itemset((i,j), b)
```

```python
img_out=(img_out-img_out.min())*255/img_out.max()
```

```python
plt.imshow(img_out,cmap=plt.get_cmap('gray'))
```
