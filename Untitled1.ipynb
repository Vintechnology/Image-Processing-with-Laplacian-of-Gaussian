# Image Processing with Laplacian of Gaussian
### Import needed libraries

```python
import numpy as np # for science computation
import imageio # load image
import matplotlib.pyplot as plt # plot and show image
```

```python
# parameters: 
# -source: the path to the image; type=String 
# -alpha: 

def apply_LoG(source, alpha):
    img = imageio.imread('sample.png', pilmode='L') # read image from current directory named 'sample.png' in gray scale
    
    height = img.shape[0]
    width = img.shape[1]
    img_out= img_out=np.zeros((height,width),dtype=np.uint8) # create an height x width array filled with zero
    laplace = (1.0/alpha) * np.array( # the Laplacian of Gaussian Kernel
            [[0, 0, 1, 0, 0],
             [0, 1, 2, 1, 0],
            [1, 2, -16, 2, 1],
             [0, 1, 2, 1, 0],
             [0, 0, 1, 0, 0]])

    # Convolution Step: sliding the kernel through image
    
    # Start from 2,2 position to height-2 , width-2
    for row in np.arange(2, height-2): 
        for column in np.arange(2, width-2):        
            sum = 0 
            for kernel_row in np.arange(-2, 3):  # for each value of the kernel, multiply it to corresponding value of image
                for kernel_column in np.arange(-2, 3): 
                    value = img.item(row+kernel_row, column+kernel_column) # get pixel value 0-255
                    weight = laplace[2+kernel_row, 2+kernel_column] #get kernel value        
                    sum = sum + (weight * value) # add to sum
            kernel_output = sum
            img_out.itemset((row,column), kernel_output) # set the result to corresponding pixel of output image
    return [img,img_out]
```

```python
img,img_out=apply_LoG('sample.png',64)
```

```python
plt.imshow(img,cmap=plt.get_cmap('gray')) # show the gray-scale original image
```
